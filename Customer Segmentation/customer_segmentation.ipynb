```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import os
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Load the data
data = pd.read_csv('Mall_Customers.csv')
```

```python
# Encode the Gender column
label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])
```

```python
#view of data
print(data.head())
```

```python
# Select relevant features for clustering
features = data[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Set the environment variable to avoid the memory leak warning
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
# Determine the optimal number of clusters using the elbow method
wcss = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# Apply K-means clustering with 5 clusters (based on elbow method)
kmeans = KMeans(n_clusters=5, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Add the cluster labels to the original data
data['Cluster'] = clusters

# Display the first few rows of the data with cluster labels
print(data.head())
```

```python
# Analyze the cluster characteristics
cluster_analysis = data.groupby('Cluster').mean()
print(cluster_analysis)
```
